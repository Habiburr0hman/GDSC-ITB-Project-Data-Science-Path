```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("Posisi_gaji.csv")
dataset
```

```python
x = dataset.iloc[:, 1:2].values
x
```

```python
y = dataset.iloc[:, 2].values
y
```

```python
#Fitting model decision tree regression
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(x, y)
```

```python
#Memprediksi hasil DTR
yPred = regressor.predict([[6.5]])
yPred
```

```python
#Visualisasi hasil DTR
xGrid = np.arange(min(x), max(x), 0.01)
xGrid
```

```python
xGrid = xGrid.reshape(len(xGrid), 1)
xGrid
```

```python
plt.scatter(x, y, color="red")
plt.plot(xGrid, regressor.predict(xGrid), color="blue")
plt.title("Decision Tree Regression")
plt.xlabel("Level Jabatan")
plt.ylabel("Gaji")
plt.show()
```
