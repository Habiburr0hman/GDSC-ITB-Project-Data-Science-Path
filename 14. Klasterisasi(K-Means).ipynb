```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("Pengunjung_mall.csv")
dataset
```

```python
x = dataset.iloc[:, [3, 4]].values
x
```

```python
#Menggunakan metode elbow untuk menentukan jumlah cluster yang tepat
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init="k-means++", random_state=42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title("Metode Elbow")
plt.xlabel("Jumlah Cluster")
plt.ylabel("WCSS") #Semakin rendah WCSS, semakin baik modelnya
plt.show()
```

```python
#Menjalankan K-Means Clustering ke dataset
kmeans = KMeans(n_clusters=5, init="k-means++", random_state=42)
y_kmeans = kmeans.fit_predict(x)
y_kmeans
```

```python
#Visualisasi hasil K-Means Clustering
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s=50, c="red", label="Cluster 1")
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], s=50, c="blue", label="Cluster 2")
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], s=50, c="green", label="Cluster 3")
plt.scatter(x[y_kmeans == 3, 0], x[y_kmeans == 3, 1], s=50, c="cyan", label="Cluster 4")
plt.scatter(x[y_kmeans == 4, 0], x[y_kmeans == 4, 1], s=50, c="magenta", label="Cluster 5")
plt.scatter(x[y_kmeans == 5, 0], x[y_kmeans == 5, 1], s=50, c="orange", label="Cluster 6")
plt.scatter(x[y_kmeans == 6, 0], x[y_kmeans == 6, 1], s=50, c="purple", label="Cluster 7")
plt.scatter(x[y_kmeans == 7, 0], x[y_kmeans == 7, 1], s=50, c="brown", label="Cluster 8")
plt.scatter(x[y_kmeans == 8, 0], x[y_kmeans == 8, 1], s=50, c="indigo", label="Cluster 9")
plt.scatter(x[y_kmeans == 9, 0], x[y_kmeans == 9, 1], s=50, c="black", label="Cluster 10")
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=150, c="yellow", label="Centroids")
            
plt.title("K-Means Clustering")
plt.xlabel("Pendapatan Tahunan (Juta Rupiah)")
plt.ylabel("Rating Pengeluaran (1-100)")
plt.legend()
plt.show()
```
