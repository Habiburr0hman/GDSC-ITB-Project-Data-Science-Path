```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("Posisi_gaji.csv")
dataset
```

```python
x = dataset.iloc[:, 1:2].values
x
```

```python
y = dataset.iloc[:, 2].values
y
```

```python
#Fitting model random forest regression
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=72, random_state=0) #semakin besar n_estimators(tree), maka semakin baik hasilnya. default: 10 - 100
regressor.fit(x, y)
```

```python
#Memprediksi hasil RFR
yPred = regressor.predict([[6.5]])
yPred
```

```python
#Visualisasi hasil RFR
xGrid = np.arange(min(x), max(x), 0.01)
xGrid
```

```python
xGrid = xGrid.reshape(len(xGrid), 1)
xGrid
```

```python
plt.scatter(x, y, color="red")
plt.plot(xGrid, regressor.predict(xGrid), color="blue")
plt.title("Random Forest Regression")
plt.xlabel("Level Posisi")
plt.ylabel("Gaji")
plt.show()
```
