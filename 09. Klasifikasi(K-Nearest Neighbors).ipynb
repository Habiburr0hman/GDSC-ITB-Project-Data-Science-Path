```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("Iklan_sosmed.csv")
dataset
```

```python
x = dataset.iloc[:, [2, 3]].values
x
```

```python
y = dataset.iloc[:, 4].values
y
```

```python
#Membagi train set dan test set
from sklearn.model_selection import train_test_split
xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size=0.25, random_state=0)
```

```python
xTrain
```

```python
xTest
```

```python
yTrain
```

```python
yTest
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
xTrain = sc.fit_transform(xTrain)
xTest = sc.fit_transform(xTest)
```

```python
xTrain
```

```python
xTest
```

```python
#Mmebuat & Fitting model K-NN
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5, metric="minkowski", p=2)
classifier.fit(xTrain, yTrain)
```

```python
#Memprediksi test set
yPred = classifier.predict(xTest)
yPred
```

```python
#Membuat confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(yTest, yPred)
cm
#Bisa dilihat bahwa akurasi model kita 64+29 = 93% (sangat baik)
```

```python
#Visualisasi hasil training set
from matplotlib.colors import ListedColormap
xSet, ySet = xTrain, yTrain
x1, x2 = np.meshgrid(np.arange(start = xSet[:, 0].min() - 1, stop = xSet[:, 0].max() + 1, step = 0.01),
                    np.arange(start = xSet[:, 1].min() - 1, stop = xSet[:, 1].max() + 1, step = 0.01)
                    )

plt.contourf(x1, x2, classifier.predict(np.array([x1.ravel(), x2.ravel()]).T).reshape(x1.shape),
            alpha = 0.75, cmap = ListedColormap(("red", "green"))
            )

plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())
for i, j in enumerate(np.unique(ySet)):
    plt.scatter(xSet[ySet == j, 0], xSet[ySet == j, 1],
               c = ListedColormap(("red", "green"))(i), label = j
               )
    
plt.title("K-NN(Training Set)")
plt.xlabel("Usia")
plt.ylabel("Estimasi Gaji")
plt.legend()
plt.show()
```

```python
#Visualisasi hasil training set dengan invers scaling(lebih lambat)
from matplotlib.colors import ListedColormap
xSet, ySet = sc.inverse_transform(xTrain), yTrain
x1, x2 = np.meshgrid(np.linspace(start = xSet[:, 0].min() - 10, stop = xSet[:, 0].max() + 10, num = 2000),
                    np.linspace(start = xSet[:, 1].min() - 35000, stop = xSet[:, 1].max() + 35000, num = 500)
                    )

plt.contourf(x1, x2, classifier.predict(sc.transform(np.array([x1.ravel(), x2.ravel()]).T)).reshape(x1.shape),
            alpha = 0.5, cmap = ListedColormap(("red", "green"))
            )

plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())
for i, j in enumerate(np.unique(ySet)):
    plt.scatter(xSet[ySet == j, 0], xSet[ySet == j, 1],
               c = ListedColormap(("red", "green"))(i), label = j
               )
    
plt.title("K-NN(Training Set)")
plt.xlabel("Usia")
plt.ylabel("Estimasi Gaji")
plt.legend()
plt.show()
```

```python
#Visualisasi hasil test set
from matplotlib.colors import ListedColormap
xSet, ySet = xTest, yTest
x1, x2 = np.meshgrid(np.arange(start = xSet[:, 0].min() - 1, stop = xSet[:, 0].max() + 1, step = 0.01),
                    np.arange(start = xSet[:, 1].min() - 1, stop = xSet[:, 1].max() + 1, step = 0.01)
                    )

plt.contourf(x1, x2, classifier.predict(np.array([x1.ravel(), x2.ravel()]).T).reshape(x1.shape),
            alpha = 0.75, cmap = ListedColormap(("red", "green"))
            )

plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())
for i, j in enumerate(np.unique(ySet)):
    plt.scatter(xSet[ySet == j, 0], xSet[ySet == j, 1],
               c = ListedColormap(("red", "green"))(i), label = j
               )
plt.title("K-NN(Test Set)")
plt.xlabel("Usia")
plt.ylabel("Estimasi Gaji")
plt.legend()
plt.show()
```
