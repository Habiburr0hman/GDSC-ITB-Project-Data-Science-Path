```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("50_Startups.csv")
dataset
```

```python
x = dataset.iloc[:, :4].values
tampilkanX = pd.DataFrame(x)
tampilkanX
x
```

```python
y = dataset.iloc[:, 4].values
y
```

```python
#Encode categorical data ==> dalam kasus ini: kolom Wilayah
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelEncoderX = LabelEncoder()
x[:, 3] = labelEncoderX.fit_transform(x[:, 3])
x
```

```python
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer(
    [("Wilayah", OneHotEncoder(), [3])], remainder="passthrough"
    )
x = np.array(ct.fit_transform(x), dtype=np.float)
x = x[:, 1:] #Menghindari jebakan dummy variable dengan membuangnya
x
```

```python
#Membagi data menjadi train set dan test set
from sklearn.model_selection import train_test_split
xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
xTrain
```

```python
xTest
```

```python
yTrain
```

```python
yTest
```

```python
#Regresi linear dan polinomial sudah ada feature scaling didalamnya, jadi kita tidak perlu scaling manual
#Membuat model multiple linear regression dari training set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(xTrain, yTrain)
```

```python
#Memprediksi hasil test set
yPred = regressor.predict(xTest)
yPred
```

```python
#Memilih model muliple regression yang paling baik dengan metode Backward Propagation secara Manual, tidak dengan fungsi
import statsmodels.api as sma
x = sma.add_constant(x)
x
```

```python
import statsmodels.api as sm
xOpt = x[:, [0, 1, 2, 3, 4, 5]]
regressorOLS = sm.OLS(endog=y, exog=xOpt).fit()
regressorOLS.summary()
```

```python
#Step: Eliminasi variabel yang "p value / p>|t|" nya sangat besar(umumnya > 0.05) berdasar OLSRegression ==> variabel indeks ke-2
xOpt = x[:, [0, 1, 3, 4, 5]]
regressorOLS = sm.OLS(endog=y, exog=xOpt).fit()
regressorOLS.summary()
```

```python
#Step: Eliminasi variabel ==> variabel indeks ke-1
xOpt = x[:, [0, 3, 4, 5]]
regressorOLS = sm.OLS(endog=y, exog=xOpt).fit()
regressorOLS.summary()
```

```python
#Step: Eliminasi variabel ==> variabel indeks ke-1
xOpt = x[:, [0, 3, 5]]
regressorOLS = sm.OLS(endog=y, exog=xOpt).fit()
regressorOLS.summary()
```

```python
x[:, [0, 3, 5]]
```

```python
#Step: Eliminasi variabel ==> variabel indeks ke-2
#Kita bisa lihat bahwa variabel yang paling signifikan terhadap y adalah kolom R&D
#Persamaan garis regresi optimal:
#profit = 49030 + 0.8543 * Biaya R&D
xOpt = x[:, [0, 3]]
regressorOLS = sm.OLS(endog=y, exog=xOpt).fit()
regressorOLS.summary()
```

```python
#Regresi Ulang & Visualisasi dengan single variabel "R&D" ==> variabel yang paling mempengaruhi
xTrain, xTest, yTrain, yTest = train_test_split(x[:, 3], y, test_size=0.2, random_state=0)
xTrain = xTrain.reshape(-1, 1)
xTest = xTest.reshape(-1, 1)
```
