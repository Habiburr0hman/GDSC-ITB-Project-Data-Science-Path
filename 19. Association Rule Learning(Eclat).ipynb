```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("optimisasi_retail.csv", header=None)
dataset
```

```python
jumBar = dataset.shape[0]
jumKol = dataset.shape[1]
transaksi = [[str(dataset.values[i, j]) for j in range(0, jumKol) if str(dataset.values[i, j]) != "nan"]
              for i in range(0, jumBar)]
transaksi
```

```python
#Membuat variabel item berisi transaksi yang tidak terduplikasi
items = list()
for t in transaksi:
    for x in t:
        if(not x in items):
            items.append(x)
items
```

```python
#Membuat list eclat yang merupakan pasangan 2 item, dengan nilai support awal 0
eclat = list()
for i in range(0, len(items)):
    for j in range(i+1, len(items)):
        eclat.append([(items[i], items[j]), 0, 0])
eclat
```

```python
#Menghitung nilai support untuk setiap pasangan transaksi
for p in eclat:
    for t in transaksi:
        if(p[0][0] in t) and (p[0][1] in t):
            p[1] += 1
    p[2] = p[1]/len(transaksi)
```

```python
#Mengubah hasil eclat menjadi dataframe dan mengurutkannya dari support tertinggi
hasilEclat = pd.DataFrame(eclat, columns = ["Rule", "Count", "Support"]).sort_values(by = "Support", ascending = False)
hasilEclat
```
